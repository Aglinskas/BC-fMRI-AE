```python
def get_ABIDE_subData(s,fldr='func',ses='ses-1',deriv='cpac_nofilt_global'):
    subFolderTemp = '../../data/ABIDE/Derivatives/{deriv}/sub-{siteID}00{subID}/{ses}/{fldr}/'
    
    siteID = df['SITE_ID_BIDS'][s]
    subID = df['SUB_ID'][s]

    subFolder = subFolderTemp.format(siteID=siteID,subID=subID,fldr=fldr,ses=ses,deriv=deriv)

    subFiles_list = os.listdir(subFolder)
    subFiles_dict = {}
    for subFile in subFiles_list:
        key = subFile.replace('sub-{siteID}00{subID}_ses-1_'.format(siteID=siteID,subID=subID),'')
        subFiles_dict.update({key : os.path.join(subFolder,subFile)})
    return subFiles_dict
```
