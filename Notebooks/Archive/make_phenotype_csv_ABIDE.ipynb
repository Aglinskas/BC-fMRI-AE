```python
deriv = 'cpac_nofilt_global'
#deriv = 'niak_nofilt_global'
#deriv = 'dparsf_nofilt_global'
#deriv = 'ccs_nofilt_global'
bids_folders = os.listdir(f'../../data/ABIDE/Derivatives/{deriv}/')
bids_folders.sort()
bids_folders = bids_folders[2:]

bids = pd.DataFrame(bids_folders,columns=['folders'])
bids['ID'] = bids.applymap(lambda val : val[-5::])
bids['ID'] = pd.to_numeric(bids['ID'])
bids = bids.sort_values(by='ID')
len(bids)
bids.head(5)
len(bids)
```

```python
n = len(df)
df['bids_folder'] = np.nan
for i in range(n):

    try:
        idx = int(np.argwhere(df['participant_id'].values[i]==bids['ID'].values))
    except:
        idx = None

    if type(idx)!=type(None):
        #print('yes')
        df['bids_folder'].iloc[i] = bids['folders'].values[idx]
```
