```python
df = pd.read_csv('../Data/YeoAtlas/Yeo2011_7Networks_N1000.split_components.FSL_MNI152_2mm.Centroid_RAS.csv')
rlbls = df['Label Name'].values[1::]
rlbls = [rlbl.replace('7Networks_','') for rlbl in rlbls]
rlbls = [rlbl.replace('_','-') for rlbl in rlbls]
rlbls[0:5]
```
