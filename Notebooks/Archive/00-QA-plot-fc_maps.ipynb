```python
import ants
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
#indir = '../Assets/fc_maps_32smooth/'
indir = '../Assets/fc_maps_orig//'
infiles = [file for file in os.listdir(indir) if file.endswith(file)]
infiles.sort()   
len(infiles)
```

```python
s = 33
im = ants.image_read(os.path.join(indir,infiles[s]))
#im = ants.image_read(os.path.join(indir,infiles[s]))
#im = im.resample_image((32,32,32,51),use_voxels=True,interp_type=4)
#im = im.smooth_image(sigma=3,sigma_in_physical_coordinates=False,FWHM=False)
#im = ants.iMath_normalize(im)
plt.imshow(im.numpy()[:,:,20,0]);plt.xticks([]);plt.yticks([]);
```

```python
nc = 10;nr=10;
plt.subplots(nr,nc,figsize=(20,20));
start_idx = 0
for i in range(nc*nr):
    plt.subplot(nr,nc,i+1)
    s = start_idx+i
    im = ants.image_read(os.path.join(indir,infiles[s])).slice_image(axis=3,idx=0)
    plt.imshow(im.numpy()[:,:,20]);plt.xticks([]);plt.yticks([]);
    plt.title(infiles[s])

plt.tight_layout()
```
